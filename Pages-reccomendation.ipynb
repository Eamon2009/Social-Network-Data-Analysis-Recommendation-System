# Pages You Might Like – Recommendation System

## Task 1: Recommendation Logic

### How It Works
- Users **like pages**.
- If two users like the **same page**, they likely share interests.
- Pages liked by one user can be **recommended to the other**.

This method is called **Collaborative Filtering**.

**Idea:**  
If two users like the same thing, they may like other pages each of them likes.

### Example
- **Amit (1)** likes: Python Hub (101), AI World (102)  
- **Priya (2)** likes: AI World (102), Data Science Daily (103)

Both like **AI World (102)**.

**Recommendations:**
- Amit → Data Science Daily (103)  
- Priya → Python Hub (101)

## Task 2: Algorithm

The function will:

1. **Map users to liked pages**
2. **Find users with common likes**
3. **Get pages liked by similar users**
4. **Rank recommendations by shared interactions**

```python
import json

# Function to load JSON data from a file
def load_data(filename):
    with open(filename, "r") as file:
        return json.load(file)

# Function to find pages a user might like based on common interests
def find_pages_you_might_like(user_id, data):
    # Dictionary to store user interactions with pages
    user_pages = {}
    for user in data["users"]:
        user_pages[user["id"]] = set(user["liked_pages"])
    
    # If the user is not found, return an empty list
    if user_id not in user_pages:
        return []
    
    user_liked_pages = user_pages[user_id]
    page_suggestions = {}
    
    for other_user, pages in user_pages.items():
        if other_user != user_id:
            shared_pages = user_liked_pages.intersection(pages)
            for page in pages:
                if page not in user_liked_pages:
                    page_suggestions[page] = page_suggestions.get(page, 0) + len(shared_pages)
    
    # Sort recommended pages based on the number of shared interactions
    sorted_pages = sorted(page_suggestions.items(), key=lambda x: x[1], reverse=True)
    return [page_id for page_id, _ in sorted_pages]
```

# Output 
 - Recommendation pages for user_id=1

```python

# Load data
data = load_data("Data_set2.json")
user_id = 1  # Example
page_recommendations = find_pages_you_might_like(user_id, data)
print(f"Pages You Might Like for User {user_id}: {page_recommendations}")
```
